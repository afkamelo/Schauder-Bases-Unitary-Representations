```python
# ------------------------------------------------------------
# Haar vs Fourier en [-3,3]: tres ejemplos y figuras en .eps
# Estilo simple de graficación
# ------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt

# ---------------- Utilidades de malla e inner products ----------------
def make_grid(a=-3.0, b=3.0, n=8192):
    t = np.linspace(a, b, n, endpoint=False)
    w = (b - a) / n
    return t, w

# ---------------- Haar en [0,1) y mapeo a [-3,3] ----------------
def haar_phi(u):
    return np.where((u >= 0.0) & (u < 1.0), 1.0, 0.0)

def haar_psi(u):
    out = np.zeros_like(u)
    out[(u >= 0.0) & (u < 0.5)] = 1.0
    out[(u >= 0.5) & (u < 1.0)] = -1.0
    return out

def phi_jk_u(u, j, k):
    return (2.0 ** (j / 2.0)) * haar_phi((2.0 ** j) * u - k)

def psi_jk_u(u, j, k):
    return (2.0 ** (j / 2.0)) * haar_psi((2.0 ** j) * u - k)

def map_to_unit(t, a=-3.0, b=3.0):
    return (t - a) / (b - a)

def haar_projection_interval(f, J=3, a=-3.0, b=3.0, ngrid=8192):
    t, w = make_grid(a, b, ngrid)
    L = b - a
    u = map_to_unit(t, a, b)
    fx = f(t)
    norm = 1.0 / np.sqrt(L)

    phi00 = norm * phi_jk_u(u, 0, 0)
    c00 = np.sum(fx * phi00) * w
    recon = c00 * phi00

    for j in range(J):
        for k in range(2 ** j):
            psi_jk = norm * psi_jk_u(u, j, k)
            d_jk = np.sum(fx * psi_jk) * w
            recon += d_jk * psi_jk

    return t, fx, recon

# ---------------- Fourier en [-3,3] (periodo 6) ----------------
def fourier_coeffs_interval(f, M=3, a=-3.0, b=3.0, ngrid=8192):
    t, w = make_grid(a, b, ngrid)
    L = b - a
    fx = f(t)
    a0 = (2.0 / L) * np.sum(fx * np.ones_like(t)) * w
    a_n = np.zeros(M + 1)
    b_n = np.zeros(M + 1)
    a_n[0] = a0
    tau = (t - a) / L
    for n in range(1, M + 1):
        cosn = np.cos(2.0 * np.pi * n * tau)
        sinn = np.sin(2.0 * np.pi * n * tau)
        a_n[n] = (2.0 / L) * np.sum(fx * cosn) * w
        b_n[n] = (2.0 / L) * np.sum(fx * sinn) * w
    return t, fx, a_n, b_n

def fourier_reconstruct_on_grid(a_n, b_n, t, a=-3.0, b=3.0):
    L = b - a
    tau = (t - a) / L
    M = len(a_n) - 1
    S = 0.5 * a_n[0] * np.ones_like(t)
    for n in range(1, M + 1):
        S += a_n[n] * np.cos(2.0 * np.pi * n * tau) + b_n[n] * np.sin(2.0 * np.pi * n * tau)
    return S

# ---------------- Fourier periódico con periodo T (CORREGIDO) ----------------
def fourier_coeffs_periodic_T(f_periodic, M=3, T=1.0, ngrid=8192, t0=0.0):
    """
    Coeficientes de Fourier de una función T-periódica.
    Integra en [t0, t0+T] y usa base cos/sin con periodo T.
    """
    t = np.linspace(t0, t0 + T, ngrid, endpoint=False)
    w = T / ngrid
    fx = f_periodic(t)

    a0 = (2.0 / T) * np.sum(fx) * w
    a_n = np.zeros(M + 1)
    b_n = np.zeros(M + 1)
    a_n[0] = a0
    tau = (t - t0) / T
    for n in range(1, M + 1):
        cosn = np.cos(2.0 * np.pi * n * tau)
        sinn = np.sin(2.0 * np.pi * n * tau)
        a_n[n] = (2.0 / T) * np.sum(fx * cosn) * w
        b_n[n] = (2.0 / T) * np.sum(fx * sinn) * w
    return a_n, b_n

def fourier_reconstruct_periodic_T(a_n, b_n, t, T=1.0, t0=0.0):
    """Reconstrucción periódica con periodo T sobre puntos t (usa extensión periódica)."""
    tau = ((t - t0) % T) / T
    M = len(a_n) - 1
    S = 0.5 * a_n[0] * np.ones_like(t)
    for n in range(1, M + 1):
        S += a_n[n] * np.cos(2.0 * np.pi * n * tau) + b_n[n] * np.sin(2.0 * np.pi * n * tau)
    return S

# ---------------- Funciones de prueba ----------------
def f_quad(t):  # no periódica
    return t ** 2

def square_wave_period1(t):
    frac = np.mod(t, 1.0)
    out = np.ones_like(frac)
    out[frac >= 0.5] = -1.0
    return out

def triangle_wave_period1(t):
    frac = np.mod(t, 1.0)
    return np.where(frac < 0.5, 4.0 * frac - 1.0, -4.0 * frac + 3.0)

# ---------------- Helper para un ejemplo completo ----------------
def ejemplo(f, titulo, nombre_base, J=5, M=3, a=-3.0, b=3.0, ngrid=8192,
            usar_fourier_periodico=False, T=1.0):
    # Haar
    t, fx, fH = haar_projection_interval(f, J=J, a=a, b=b, ngrid=ngrid)

    # Fourier
    if usar_fourier_periodico:
        # Coeficientes en su periodo natural T y reconstrucción periódica en [-3,3]
        a_n, b_n = fourier_coeffs_periodic_T(f, M=M, T=T, ngrid=ngrid, t0=0.0)
        fF = fourier_reconstruct_periodic_T(a_n, b_n, t, T=T, t0=0.0)
    else:
        # Periodización del trozo en [-3,3] (periodo L=6)
        t2, fx2, a_n, b_n = fourier_coeffs_interval(f, M=M, a=a, b=b, ngrid=ngrid)
        fF = fourier_reconstruct_on_grid(a_n, b_n, t2, a=a, b=b)

    # --------- Gráfica 1: función + Haar + Fourier ---------
    plt.figure(figsize=(10,6))
    plt.plot(t, fx, 'r', label='f(t)')
    plt.plot(t, fH, 'b', label='Haar')
    plt.plot(t, fF, 'g', label='Fourier')
    plt.grid()
    plt.title(titulo)
    plt.xlabel('t')
    plt.ylabel('f(t)')
    plt.legend()
    plt.savefig(f"{nombre_base}.eps", format="eps")
    plt.show()

    # --------- Gráfica 2: errores puntuales ---------
    plt.figure(figsize=(10,6))
    errH = np.abs(fx - fH)
    errF = np.abs(fx - fF)
    plt.plot(t, errH, 'b', label='|f - Haar|')
    plt.plot(t, errF, 'g', label='|f - Fourier|')
    plt.grid()
    plt.title(f"$L^1$ norm of error — {titulo}")
    plt.xlabel('t')
    plt.ylabel('|e(t)|')
    plt.legend()
    plt.savefig(f"{nombre_base}_errores.eps", format="eps")
    plt.show()

# ---------------- Ejecutar 3 ejemplos ----------------
if __name__ == "__main__":
    J = 6
    M = 3
    a, b = -3.0, 3.0
    ngrid = 8192

    # 1) y = x^2 en [-3,3] (Fourier con periodo L=6 del intervalo)
    ejemplo(f_quad,
            "Quadratic Function",
            "ej1_cuadratica",
            J=J, M=M, a=a, b=b, ngrid=ngrid,
            usar_fourier_periodico=False)

    # 2) Onda rectangular T=1 (usar Fourier con periodo T=1)
    ejemplo(square_wave_period1,
            "Rectangular Wave",
            "ej2_onda_rectangular",
            J=J, M=M, a=a, b=b, ngrid=ngrid,
            usar_fourier_periodico=True, T=1.0)

    # 3) Onda triangular T=1 (usar Fourier con periodo T=1)
    ejemplo(triangle_wave_period1,
            "Triangular Wave",
            "ej3_onda_triangular",
            J=J, M=M, a=a, b=b, ngrid=ngrid,
            usar_fourier_periodico=True, T=1.0)
```
